```

N = 500000   

# For autompg.csv
mpgmean = autompg.agg({'mpg': 'mean'})
mpgstd = autompg['mpg'].std()
random_mpgmean = np.random.normal(loc=mpgmean, scale=mpgstd, size=N)

displacementmean = autompg.agg({'displacement': 'mean'})
displacementstd = autompg['displacement'].std()
random_displacement = np.random.normal(loc=mpgmean, scale=mpgstd, size=N)

horsepowermean = autompg.agg({'horsepower': 'mean'})
horsepowerstd = autompg['horsepower'].std()
random_horsepower = np.random.normal(loc=mpgmean, scale=mpgstd, size=N)
 
weightmean = autompg.agg({'weight': 'mean'})
weightstd = autompg['weight'].std()
random_weight = np.random.normal(loc=mpgmean, scale=mpgstd, size=N)
 
accelarationmean = autompg.agg({'accelaration': 'mean'})
accelarationstd = autompg['accelaration'].std()
random_acceleration = np.random.normal(loc=accelarationmean, scale=accelarationstd, size=N)

# For linearregression.csv
dvs1mean = linerregression.agg({'dvs1': 'mean'})
dvs1std = linerregression['dvs1'].std()
random_dvs1mean = np.random.normal(loc=dvs1mean, scale=dvs1std, size=N)

ivs1mean = linerregression.agg({'ivs1': 'mean'})
ivs1std = linerregression['ivs1'].std()
random_ivs1mean = np.random.normal(loc=ivs1mean, scale=ivs1std, size=N)

ivs2mean = linerregression.agg({'ivs2': 'mean'})
ivs2std = linerregression['ivs2'].std()
random_ivs2mean = np.random.normal(loc=ivs2mean, scale=ivs2std, size=N)
 
ivs3mean = linerregression.agg({'ivs3': 'mean'})
ivs3std = linerregression['ivs3'].std()
random_ivs3mean = np.random.normal(loc=ivs3mean, scale=ivs3std, size=N)
```

```
from pyspark.mllib.regression import LabeledPoint
import pyspark.mllib
import pyspark.mllib.regression
from pyspark.sql.functions import *
import numpy as np
from pyspark.mllib.linalg import Vectors

# Created RDD FILE of linearregression , used for prediction without labeled point
linerregression_f = spark.createDataFrame(linerregression)

# Used for creating Labeled point and Prediction
linerregression_frdd=linerregression_f.rdd
print("RDD File of linear regression")
linerregression_frdd.take(5)
```

<center> <h2> Without Labeled Point</h2></center>

```
# Vectorization of Data of linearregression
from pyspark.ml.feature import VectorAssembler
linerregression_f = linerregression_f.withColumnRenamed('dvs1', "target")  

vectorAssembler = VectorAssembler(inputCols = ['ivs1','ivs2','ivs3',], outputCol = 'features')
linerregression_df_assembled = vectorAssembler.transform(linerregression_f)
linerregression_df_assembled = linerregression_df_assembled.select(['features', 'target'])
linerregression_df_assembled.show(5, truncate=False)
```

```
# Dividing the Dataset into training and test data
(train_data, test_data) = linerregression_df_assembled.randomSplit([0.7, 0.3])
print("Train count:", train_data.count())
print("Test count:", test_data.count())
```

```
# Developing a Linear Regression Model 
from pyspark.ml.regression import LinearRegression
rf1 = LinearRegression(featuresCol="features", labelCol="target", maxIter=100,regParam=0.0, elasticNetParam=0.0,aggregationDepth=2,fitIntercept=True, standardization=True)

# Training the Data
modellr = rf1.fit(train_data)
print("Coefficients: " + str(modellr.coefficients))
print("Intercept: " + str(modellr.intercept))
print("")
print("Training Summary")

trainingSummary = modellr.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
print("objectiveHistory: %s" % str(trainingSummary.objectiveHistory))
trainingSummary.residuals.show()
```

```
# For Deleteing  an Existing Path
dbutils.fs.rm('/tmp/mllib-persistence-example/modellr', True)

# Saving an Model
from pyspark.ml.regression import LinearRegression

# let modellr be a Regression Model
basePath = "/tmp/mllib-persistence-example"
modellr.save(basePath + "/modellr")

# Loading an Model
from pyspark.ml.regression import LinearRegressionModel
loaded_model_lr=LinearRegressionModel.load('/tmp/mllib-persistence-example//modellr')
```

```
#Prediction on Testing Data using Loaded Model and Evaluation
from pyspark.ml.evaluation import RegressionEvaluator
evaluator = RegressionEvaluator(labelCol="target", predictionCol="prediction", metricName="rmse")

predictions_lr = loaded_model_lr.transform(test_data)

predictions_lr.select("features","target", "prediction").show(10)
rmse1 = evaluator.evaluate(predictions_lr)
print("Linear Regression RMSE:", rmse1)
```

<h4>Compare the lasso regression and ridge regression models and provide your commentary.<h4>

Lasso regression is a type of linear regression that uses shrinkage. Shrinkage is where data values are shrunk towards a central
point, like the mean.

Ridge regression is a method of estimating the coefficients of multiple-regression models in scenarios where linearly independent variables are highly correlated.

The main difference between Ridge and LASSO Regression is that if ridge regression can shrink the coefficient close to 0 so that all predictor variables are retained. Whereas LASSO can shrink the coefficient to exactly 0 so that LASSO can select and discard the predictor variables that have the right coefficient of 0.

During training, Lasso introduced a new hyperparameter, alpha, the coefficient to penalize weights. Ridge takes a step further and penalizes the model for the sum of squared
value of the weights.
